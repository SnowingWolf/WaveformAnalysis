```python
from DAQAnalyzer import DAQAnalyzer, DAQRun
import json

# 1. åˆå§‹åŒ–å¹¶æ‰«æ
analyzer = DAQAnalyzer()
analyzer.scan_all_runs()
print(f"å·²æ‰«æ {len(analyzer.runs)} ä¸ªè¿è¡Œï¼Œæ€»å¤§å°: {analyzer.format_size(analyzer.total_bytes)}")

# 2. æ˜¾ç¤ºç»Ÿè®¡æŠ¥è¡¨
analyzer.display_overview()
analyzer.display_summary()

# 3. æ·±åº¦åˆ†ææŒ‡å®šè¿è¡Œ
run_name = "50V_OV_circulation_20thr"
analyzer.display_run_channel_details(run_name)

# 4. è‡ªå®šä¹‰æå–æ•°æ®
run = analyzer.get_run(run_name)
if run:
    print(f"\nè¿è¡Œè¯¦æƒ…: {run.run_name} ({analyzer.format_size(run.total_bytes)})")
    stats = run.get_channel_summary()
    for ch, s in sorted(stats.items()):
        time_info = f"{s['duration_s']:.2f} s" if s["start_time_ps"] else "æ— æ—¶é—´æˆ³"
        print(f"  CH{ch}: {s['file_count']} æ–‡ä»¶ | æ—¶é•¿: {time_info}")

# 5. ä¿å­˜å¹¶éªŒè¯ç»“æœ
output = analyzer.save_to_json("daq_analysis_results.json", include_file_details=True)
if output:
    with open(output, "r", encoding="utf-8") as f:
        data = json.load(f)
    print(f"\nâœ“ ç»“æœå·²ä¿å­˜å¹¶éªŒè¯: {len(data['runs'])} ä¸ªè¿è¡Œæ•°æ®")
```
